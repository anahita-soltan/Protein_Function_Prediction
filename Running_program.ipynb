**Compare to Baseline**

```python
import pandas as pd

# Paths
baseline_f = "/content/drive/MyDrive/Bio_Data/biological_data_pfp/baseline/baseline_results/evaluation_best_f.tsv"
ours_f     = "/content/drive/MyDrive/Bio_Data/output/cafa_val_results/evaluation_best_f.tsv"

base = pd.read_csv(baseline_f, sep="\t")
ours = pd.read_csv(ours_f,     sep="\t")

def row_for(df, filename, label):
    sub = df[df["filename"] == filename]
    # map GO namespace → nice short name
    ns_map = {
        "molecular_function": "MF",
        "biological_process": "BP",
        "cellular_component": "CC",
    }
    return (
        sub.assign(ns_short=sub["ns"].map(ns_map))
           .set_index("ns_short")["f"]  # macro F-score column from CAFA
           .rename(label)
    )

rows = []
rows.append(row_for(base, "naive_submission.tsv", "Naive"))
rows.append(row_for(base, "ipr_submission.tsv",   "InterPro"))
rows.append(row_for(ours, "ours_val.tsv",         "Ours (Emb+BLAST)"))

cmp_df = pd.concat(rows, axis=1).T[["MF", "BP", "CC"]]

print(cmp_df.to_markdown(floatfmt=".3f"))
cmp_df
```
